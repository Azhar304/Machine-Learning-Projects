```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./train.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
numeric_df.head()
```

```python
numeric_df.isnull().sum()
```

```python

numeric_df = numeric_df.drop(columns=['MasVnrArea',], axis=1)
numeric_df['GarageYrBlt'] = numeric_df['GarageYrBlt'].fillna(numeric_df['GarageYrBlt'].mean())
numeric_df['LotFrontage'] = numeric_df['LotFrontage'].fillna(numeric_df['LotFrontage'].mean())
```

```python

X=numeric_df.drop(columns='SalePrice')


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_df = X.copy()

scaled_df[X.columns] = scaler.fit_transform(scaled_df[X.columns])

X_Scaled=scaled_df
Y=numeric_df['SalePrice']
X_Scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_Scaled, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

# Hyperparameter grid
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [2, 5, 10],
    'max_features': ['sqrt', 'log2'],
    'max_leaf_nodes': [10, 20, 50],
    'ccp_alpha': [0.01, 0.05, 0.1],
}

# Set up GridSearchCV
grid_search = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid, cv=5, n_jobs=-1)

# Fit model
grid_search.fit(X_train, Y_train)

# Get the best parameters
best_params = grid_search.best_params_
print(best_params)


```

```python
model = DecisionTreeRegressor(
    ccp_alpha=0.01,            
    max_depth=10,              
    max_features='sqrt',       
    max_leaf_nodes=50,         
    min_samples_leaf=10,      
    min_samples_split=10      
    
)


model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_train_pred = model.predict(X_train)
train_rmse = mean_squared_error(Y_train, y_train_pred, squared=False)
print(f"Training RMSE is {train_rmse:.2f}")

y_test_pred = model.predict(X_test)
test_rmse = mean_squared_error(Y_test, y_test_pred, squared=False)
print(f"Testing RMSE is {test_rmse:.2f}")
```

```python
test_data = pd.read_csv('./test.csv')

numeric_test = test_data.select_dtypes(include=['float64', 'int64'])
numeric_test = numeric_test.drop(columns=['MasVnrArea'], axis=1)
numeric_test['GarageYrBlt'] = numeric_test['GarageYrBlt'].fillna(numeric_test['GarageYrBlt'].mean())
numeric_test['LotFrontage'] = numeric_test['LotFrontage'].fillna(numeric_test['LotFrontage'].mean())

numeric_test.head()
```

```python
scaled_test = numeric_test.copy()
scaled_test[numeric_test.columns] = scaler.transform(scaled_test[numeric_test.columns])
print(scaled_test.isnull().sum())
```

```python
import numpy as np

scaled_test.replace({0: np.nan, 'NA': np.nan}, inplace=True)

scaled_test.isnull().sum()


```

```python

test_ids = test_data['Id']

test_predictions = model.predict(scaled_test)

submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': test_predictions
})

submission.to_csv('submission3.csv', index=False)

print("Submission file 'submission.csv' created successfully.")
```
