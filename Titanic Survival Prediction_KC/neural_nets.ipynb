```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
test_data = pd.read_csv('./test.csv')
df=pd.read_csv('./train.csv')

df.head()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Embarked']=df['Embarked'].fillna(method='bfill')
print(f"Missing age values: {df['Age'].isnull().sum()},\nMissing Embarked values: {df['Embarked'].isnull().sum()}")
df.describe()
```

```python
df=df.drop(columns=['PassengerId','Name', 'Ticket','Cabin','Parch'], axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df['Embarked']=le.fit_transform(df['Embarked'])



df.head(5)
```

```python
X=df.drop(columns='Survived', axis=1)
Y=df['Survived']

X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
from keras.layers import Dense
from keras.models import Sequential


model=Sequential([
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
from keras.losses import binary_crossentropy
from keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.001),loss=binary_crossentropy, metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=20, validation_split=0.2, batch_size=24)
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(f"Test Accuracy: {accuracy:.2f}")
```

```python
predictions = model.predict(X_test)
predicted_classes = (predictions > 0.5).astype(int)  # Convert probabilities to binary (0 or 1)
print(predicted_classes[:10])
```

```python


test_data['Age'] = test_data['Age'].fillna(df['Age'].mean())
test_data['Embarked'] = test_data['Embarked'].fillna(method='bfill')

test_data = test_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin','Parch'], axis=1)


le=LabelEncoder()
test_data['Sex']=le.fit_transform(test_data['Sex'])
test_data['Embarked']=le.fit_transform(test_data['Embarked'])
ss=StandardScaler()
test_data[['Age','Fare','Pclass','Sex','Embarked','SibSp']]= ss.fit_transform(test_data[['Age','Fare','Pclass','Sex','Embarked','SibSp']])
```

```python
predictions = model.predict(test_data)

# Convert predictions to binary values (0 or 1) using a threshold of 0.5
binary_predictions = (predictions > 0.5).astype(int)
```

```python
original_test_data=pd.read_csv('./test.csv')
submission = pd.DataFrame({
    'PassengerId': original_test_data['PassengerId'],
    'Survived': binary_predictions.ravel()  # Ensure it's 1D
})

# Save the predictions to a CSV file
submission.to_csv('submission_nets.csv', index=False)
```
