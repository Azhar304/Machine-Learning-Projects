```python
df[['Deck', 'Num', 'Place']]=df['Cabin'].str.split('/', expand=True)
df = df.drop(columns='Cabin')
df.head()


```

```python
df['Spent'] = df[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].sum(axis=1)
df['Rich']=df['Spent']>2000

df = df.drop(columns=['Name'])

df['Elite'] = (df['VIP'] & df['Rich'])

df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
for col in df.columns:  
    if df[col].isnull().sum() > 0:  
        df[col] = df[col].fillna(df[col].mode()[0]) 


df.isnull().sum().sort_values(ascending=True)
```

```python
df[['Transported','CryoSleep','Rich','VIP']]=df[['Transported','CryoSleep','Rich','VIP']].astype(int)

from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
cols_to_encode=['HomePlanet','Destination','Deck','Place']
for col in cols_to_encode:
    df[col]=encoder.fit_transform(df[col])

    
df.dtypes
```
